```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
X,y = make_blobs(n_samples=1000, centers=3, n_features=2, cluster_std=1, center_box=(-7, 7))
```

```python
df = pd.DataFrame({'feature1': X[:, 0],'feature2': X[:, 1], 'label': y})
```

```python
df.to_csv('dataset_kmeans.csv', sep=';', encoding='utf-8', index=None)
```
